# Object Selection

```python
# Function to remove items below the threshold from a list
def remove_below_threshold(row, column_name, threshold):
    filtered_list = [item for item in row[column_name] if item >= threshold]
    row[column_name] = filtered_list
    row['nTau'] = len(filtered_list)
    return row

# Column name for which to remove items below the threshold
column_name = 'Tau_pt'
# Threshold for filtering
threshold = 20.0

# Apply the function to each row
df = df.apply(remove_below_threshold, args=(column_name, threshold), axis=1)

df
```
